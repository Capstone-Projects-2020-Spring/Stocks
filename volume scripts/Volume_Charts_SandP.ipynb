```python
days = ['M', 'T', 'W', 'Th', 'F']
plt.clf()
for stock in stock_list:
    print(stock)
    df = pd.read_csv("moving average tests/stock_csv_avg_test/" + stock + '.csv') 
    df= df[['5. volume']]
    volumes = df.iloc[0:5].values
    result = volumes.ravel() 
    print(result)
    
    plt.bar(days, result, color = 'violet')
    plt.title(stock + ': Average Volume per Day in Past Month')
    plt.xlabel('Days')
    plt.ylabel('Volume')
    #plt.show()
    
    
    # adding to plots_test folder
    plt.savefig('SP100_volume_charts/' + stock + '.png')
    plt.clf()
    # clears plot so that plots do not overlap on one plot
    #plt.clf()
```
