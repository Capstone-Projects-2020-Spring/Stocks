```python
#Install the dependencies
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
```

```python
arr = []
with open("NASDAQ_ex.txt", "r") as filestream:
    for line in filestream:
        currentline = line.split(',')
        arr.append(currentline)
    stock_list = np.array(arr[0])


days = ['M', 'T', 'W', 'Th', 'F']
plt.clf()
for stock in stock_list:
    
    print(stock)
    df = pd.read_csv("Stock_CSV/" + stock + '.csv') 
    df= df[['5. volume']]
    volumes = df.iloc[0:5].values
    result = volumes.ravel() 
    print(result)
    
    plt.bar(days, result, color = 'violet')
    plt.title(stock + ': Average Volume per Day in Past Month')
    plt.xlabel('Days')
    plt.ylabel('Volume')
    #plt.show()
    
    
    # adding to plots_test folder
    plt.savefig('VOLUME_CHARTS/NASDAQ_volume_charts/' + stock + '.png')
    plt.clf()
    # clears plot so that plots do not overlap on one plot
    #plt.clf()
```
