# Testing to see if other stocks also show smooth moving avg

```python
#Install the dependencies
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
```

```python
# Storing FB stock data into our dataframe df
df = pd.read_csv("stock_csv_avg_test/AAPL.csv") 
```

```python
## let's take the dataframe df and only take look at '4. close' frame
df= df[['4. close']]
```

```python
df.head()
```

```python
# lets find the Moving Average for the last 30 days.
indextest = df.iloc[:30]
```

```python
# These are all the prices we are looking at
print(indextest)
# Just checking that we get 30 prices
print('length:', len(indextest))
```

```python
# converting dataframe df of the first 30 rows to an array, closing_prices
closing_prices = df.iloc[0:30].values
print(closing_prices)
type(closing_prices)

# making sure its array
print(closing_prices[0])

#finding length of array
print('length:',len(closing_prices))
```

```python
# empty list
moving_avg = []

# interval we want to take the average from
interval = 5

# store length of array
len_prices = len(closing_prices)
```

```python
# iterate through the prices and find the average of 
for i in range(len_prices):
    if i==(len_prices-interval-1):
        break
    ## summing up prices of intervals
    sum_prices = 0 
    
    for j in range(interval): 
        print(closing_prices[i+j])
        sum_prices += closing_prices[i+j]
        
    ## appending to list averages of intervals
    moving_avg.append(sum_prices/interval) 
    print(sum_prices)
    print('\n')     
            
print(moving_avg)
```

```python
#lst = lst[:len(lst)-n]
# only keeping first n - interval rows for the graph - same amount of days match 
delete_intvl = closing_prices[:len(closing_prices)-interval]
print(delete_intvl)
```

```python
plt.plot(moving_avg)
plt.plot(delete_intvl)
plt.title('Moving Average Mirror for FB')
plt.xlabel('Days')
plt.ylabel('Prices')
```
