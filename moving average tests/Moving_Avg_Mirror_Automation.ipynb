```python
#Install the dependencies
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
```

```python
## stock list to iterate through
stock_list = ['AAPL','ABBV','ABT','ACN','AGN','AIG','ALL','AMGN','AMZN','BA','BAC','BIIB','BK','BKNG',
              'BMY','BRK.B','C','CAT','CHTR','CMCSA','COF','COST','CSCO','CVS','DD','DHR','DIS',
                'DOW','DUK','EXC','F','FB','FDX','GD','GILD','GM','GOOG','GOOGL','GS','HON','IBM',
                'INTC','JNJ','JPM','KMI','KO','LLY','LMT','LOW','MCD','MDLZ','MDT','MET','MMM','MRK',
                'MS','MSFT','NEE','NFLX','NVDA','ORCL','OXY','PEP','PFE','PM','PYPL','QCOM','RTN','SBUX',
                'SO','SPG','T','TGT','TMO','UNH','UNP','UPS','USB','UTX','VZ','WBA','WFC','WMT','XOM']

# empty list to put in moving averages
moving_avg = []

# interval we want to take the average from -->we call this mirror<--
interval = 5
```

```python
for stock in stock_list:
    print(stock)
    df = pd.read_csv("stock_csv_avg_test/" + stock + '.csv') 
    df= df[['4. close']]
    closing_prices = df.iloc[0:30].values
    len_prices = len(closing_prices)
    
    for i in range(len_prices):  
        if i==(len_prices-interval-1):
            break
        sum_prices = 0 
        
        for j in range(interval): 
            print(closing_prices[i+j])
            sum_prices += closing_prices[i+j]
        
        moving_avg.append(sum_prices/interval) 
        print('sum:',sum_prices)
        #print('\n') 
        print('moving avg by computer:',moving_avg)
        print('\n')
    
    #lst = lst[:len(lst)-n]
    # only keeping first n - interval rows for the graph - same amount of days match 
    delete_intvl = closing_prices[:len(closing_prices)-interval]
    print(delete_intvl)
    
    #plotting moving average mirror vs actual prices
    plt.plot(moving_avg, color ='mediumvioletred')
    plt.plot(delete_intvl, color = 'darkolivegreen')
    plt.title('Moving Average Mirror for '+ stock)
    plt.xlabel('Days')
    plt.ylabel('Prices')
    
    # adding to plots_test folder
    plt.savefig('MOVING_AVG_CHARTS/plots_MACD_S&P100/' + stock + '.png')
    
    # clears plot so that plots do not overlap on one plot
    plt.clf()
   
    
    moving_avg.clear()
    
```
